# First Component: Class LabeledList

Part I: This is the first Part of the TableTools module. 

```python
class LabeledList():
    
    def __init__(self, data = None, index = None):
        self.values = data
        if index == None:
            self.index = [i for i in range(0, len(self.values))]
        else:
            self.index = index

    def __str__(self):
        s = "";
        for i in range(0, len(self.values)):
            s += "%s %s\n" %(self.index[i], self.values[i])
        return s

    def __repr__(self):
        return str(self)
    def __getitem__(self, key_list):
        if isinstance(key_list, LabeledList):
            abc = key_list.values
            ind = []
            val = []
            for value in range (0, len(self.values)):
                if self.index[value] in abc:
                    ind.append(self.index[value])
                    val.append(self.values[value])
            return LabeledList(val, ind)
        elif isinstance(key_list, list):
            if isinstance(key_list[0],type(self.index[0])):
                ind = []
                val = []
                for value in range (0, len(self.values)):
                    if self.index[value] in key_list:
                        ind.append(self.index[value])
                        val.append(self.values[value])
                return LabeledList(val, ind)
            if isinstance(key_list[0], bool):
                ind = []
                val = []
                for value in range(0, len(self.values)):
                    if key_list[value] == True:
                        ind.append(self.index[value])
                        val.append(self.values[value])
                return LabeledList(val, ind)
        elif isinstance(key_list, type(self.index[0])):
            counter = 0
            ind = []
            val = []
            for value in range (0, len(self.values)):
                if self.index[value] == key_list:
                    ind.append(self.index[value])
                    val.append(self.values[value])
                    counter+=1
            if counter > 1:
                return LabeledList(val, ind)
            else:
                return val[0]
    def __setitem__(self, key, value):
        for i in range (0, len(self.values)):
            if self.index[i] == key:
                self.values[i] = value
        return self
    def __iter__(self):
        yield from self.values
    def __eq__(self, scalar):
        bc = [] 
        for i in range (0, len(self.values)):
            if self.values[i] == scalar:
                bc.append(True)
            else:
                bc.append(False)
        return LabeledList(bc, self.index)
    def __ne__(self, scalar):
        cd = []
        for i in range (0, len(self.values)):
            if self.values[i] != scalar:
                cd.append(True)
            else:
                cd.append(False)
        return LabeledList(cd, self.index)  
    def __gt__(self, scalar):
        ef = []
        for i in range (0, len(self.values)):
            if self.values[i] > scalar:
                ef.append(True)
            else:
                ef.append(False)
        return LabeledList(ef, self.index)
    def __lt__(self, scalar):
        fg = []
        for i in range (0, len(self.values)):
            if self.values[i] < scalar:
                fg.append(True)
            else:
                fg.append(False)
        return LabeledList(fg, self.index)
    def map(self, f):
        return LabeledList([f(x) for x in self.values], self.index)
        
    
ll = LabeledList(['A++', 'A++', 'C', 'P'], [1, 12, 7, 7])

```

# Second Component: Class Table

Part I: This is the second Part of the TableTools module. 

```python
class Table():
    
    def __init__(self, data, index=None, columns=None):
        self.values = data
        if index == None:
            self.index = [i for i in range(0, len(self.values))]
        else:
            self.index = index
        if columns == None:
            self.columns = [i for i in range(0, len(self.values[0]))]
        else:
            self.columns = columns
    
    def __inittwo__(self):
        self.rows =  []# list of row objects, we assume if order of id
        self.nameMap = {} # for faster direct lookup for row by name
    
    def addRow(self, row):
        self.rows.append(row)
        self.nameMap[row['name']] = row
    
    def getRow(self, name):
        return self.nameMap[name]
    
    def __str__(self):
        s = "";
        for i in range(0, len(self.values)):
            s += "%s %s\n %s\n" %(self.index[i], self.values[i], self.columns[i])
        return s
    
    def __str__(self):
        top_line = ' ' * 7
        for num in self.columns:
            top_line += '{}'.format(f'{num:>7}')
        top_line += '\n'
        body = ''
        for y in range(len(self.index)):
            body += '{}'.format(f'{self.index[y]:>7}')
            for x in range(len(self.columns)):
                body += '{}'.format(f'{self.values[y][x]:>7}')
            body += '\n'
        output_table = top_line + body
        return output_table
    
    def __repr__(self):
        return str(self)
    
    def __getitem__(self, col_list):
        if isinstance(col_list, LabeledList):
            col_list = col_list.values
        
        if isinstance(col_list, list):
            
            if isinstance(col_list[0], bool):
                ind = []
                val = []
                for value in range(0, len(self.index)):
                    if col_list[value] == True:
                        ind.append(self.index[value])
                        val.append(self.values[value])
                return Table(val, ind, self.columns)
            
            elif isinstance(col_list[0], type(self.columns[0])):            
                val = []
                for i in self.index:
                    val.append([])
                col = []
                for i in col_list:
                    for j in range(len(self.columns)):
                        if self.columns[j] == i:
                            for k in range(len(self.index)):
                                val[k].append(self.values[k][j])
                            col.append(self.columns[j])
                return Table(val, self.index, col)
        
        elif isinstance(col_list, type(self.index[0])):
            counter = 0
            val = []
            for i in self.index:
                val.append([])
            col = []
            for j in range(len(self.columns)):
                if self.columns[j] == col_list:
                    for k in range(len(self.index)):
                        val[k].append(self.values[k][j])
                    col.append(self.columns[j])
                    counter += 1
            if counter > 1:
                return Table(val, self.index, col)
            else:
                s = ""
                for i in range(len(self.index)):
                    s += '{}'.format(f'{str(self.index[i]):>7}')
                    s += '{}'.format(f'{str(val[i][0]):>7}')
                    s += "\n"
                return(s)
    
    def head(self, n):
        ttop_line = ' ' * 7
        for num in self.columns:
            ttop_line += '{}'.format(f'{num:>7}')
        ttop_line += '\n'
        bbody = ''
        for y in range(0, n):
            bbody += '{}'.format(f'{self.index[y]:>7}')
            for x in range(len(self.columns)):
                bbody += '{}'.format(f'{self.values[y][x]:>7}')
            bbody += '\n'
        ooutput_table = ttop_line + bbody
        return ooutput_table
    
    def tail(self, n):
        tttop_line = ' ' * 20
        for num in self.columns:
            tttop_line += '{}'.format(f'{num:>7}')
        tttop_line += '\n'
        bbbody = ''
        for y in range((len(self.index)-n), len(self.index)):
            bbbody += '{}'.format(f'{self.index[y]:>7}')
            for x in range(len(self.columns)):
                bbbody += '{}'.format(f'{self.values[y][x]:>7}')
            bbbody += '\n'
        oooutput_table = tttop_line + bbbody
        return oooutput_table
    
    def shape(self):
        bp = len(self.index)
        ap = len(self.columns)
        sp = "(%s, %s)" %(bp, ap)
        return sp
    
    def __eq__(self, other):
        if self.values != other.values:
            return False
        if self.index != other.index:
            return False
        if self.columns != other.columns:
            return False
        return True
    def __ne__(self, other):
        return not self == other 
```

```python
def read_csv(file_name):
    indexx = []
    columnn = []
    valuee = []
    f = open(file_name, "r")
    ss = 0
    for i in f:
        #print(i)
        g = i.split(",")
        #print(g)
        if ss == 0:
            columnn = g[1:]
        else:
            indexx.append(g[0])
            valuee.append(g[1:])
        ss+=1
    return Table(valuee, indexx, columnn)
```
